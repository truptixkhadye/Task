```python
df = pd.read_excel('sample_Attendance_data.xlsx')
df.head()
```

```python
df.rename(columns = {'AtliQ':'Employee_Code', 'Unnamed: 1':'Name'}, inplace=True)
df.drop(0, inplace = True)
```

```python
df.drop(df.columns[-1],axis=1, inplace = True)
```

##Calculate the total days in the month in which employees are present (in %)

```python
present = df.isin(['P']).sum(axis=1)
present_perc = (present/30)*100
```

```python
df['Present%'] = present_perc
df.head()
```

```python
PresentPercentage = df[['Name', 'Present%']]
PresentPercentage
```

##Find Sick Leave(in%)


```python
sick = df.isin(['SL']).sum(axis=1)
sick_perc = (sick/30)*100
```

```python
df['Sick%'] = sick_perc
SickPercentage = df[['Name', 'Sick%']]
SickPercentage
```

##Find the employee name with max, min days of attendance.

```python
#Maximum attendance considering only "P" from Attendance Key
Max = df['Present%'].max()
Max_Attendance = df['Name'].where(df['Present%'] == Max).dropna()
Max_Attendance
```

```python
#Minimum attendance considering only "P" from Attendance Key
Min = df['Present%'].min()
Min_Attendance = df['Name'].where(df['Present%'] == Min).dropna()
Min_Attendance
```

```python
wfh = df.isin(['WFH']).sum(axis=1)
df['total_wfh'] = (wfh/30)*100 #in(%)
WFH = df[['Name', 'total_wfh']] 
WFH.head()
```

```python
#Maximum attendance considering "P" & "WFH" from Attendance Key
Total = (df['Present%'] + df['total_wfh'])
total_max = Total.max()
Total_Max_Attendance = df['Name'].where(Total == total_max).dropna()
Total_Max_Attendance
```

```python
#Minimum attendance considering "P" & "WFH" from Attendance Key
Total = (df['Present%'] + df['total_wfh'])
total_min = Total.min()
Total_Min_Attendance = df['Name'].where(Total == total_min).dropna()
Total_Min_Attendance
```
